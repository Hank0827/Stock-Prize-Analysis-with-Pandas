** Create a new dataframe called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

** We can use pandas pct_change() method on the Close column to create a new dataframe representing this return value. Use .groupby().**

```python
df.head()
```

```python
df3 = df2[['Company','close','date']]
boa = df3[df3['Company']=='BOA']['close'].pct_change()
citi = df3[df3['Company']=='Citi']['close'].pct_change()
gold = df3[df3['Company']=='GoldmanSachs']['close'].pct_change()
jp = df3[df3['Company']=='JPMorgan']['close'].pct_change()
ms =df3[df3['Company']=='MorganStanley']['close'].pct_change()
wf = df3[df3['Company']=='WellsFargo']['close'].pct_change()
returns = pd.DataFrame([boa,citi,gold,jp,ms,wf],index=['BOA','Citi','GoldmanSachs','JPMorgan','MorganStanley','WellsFargo']).transpose()
len(df.index)
returns.set_index(df.index, inplace=True)
```

```python
returns
```

** Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. Did anything significant happen that day?**

```python
d=[]
p=[]
c = ['BOA','Citi','GoldmanSachs','JPMorgan','MorganStanley','WellsFargo']
for i in c:
    maxdate=returns[i].idxmax()
    mindate=returns[i].idxmin()
    maxprice=returns[i].max()
    minprice=returns[i].min()
    d.append(maxdate)
    d.append(mindate)
    p.append(maxprice)
    p.append(minprice)
```

```python
Company = ['BOA','BOA','Citi','Citi','GoldmanSachs','GoldmanSachs','JPMorgan','JPMorgan','MorganStanley','MorganStanley','WellsFargo','WellsFargo']
e = ['Best','Worst','Best','Worst','Best','Worst','Best','Worst','Best','Worst','Best','Worst']
hier_index = list(zip(Company,e))
hier_index = pd.MultiIndex.from_tuples(hier_index,names=('Company','Condition'))
pd.DataFrame({'Date':d,'Return':p},index=hier_index,columns=['Date','Return'])
```

### Moving Averages

** Please derive the moving averages for these stocks in the year 2016. Use .rolling() in pandas to get the rolling average calculation. ** 



```python
roll_boa = BAC['close'].rolling(window=30).mean()
roll_citi = C['close'].rolling(window=30).mean()
roll_gs = GS['close'].rolling(window=30).mean()
roll_jpm =JPM['close'].rolling(window=30).mean()
roll_ms = MS['close'].rolling(window=30).mean()
roll_wfc = WFC['close'].rolling(window=30).mean()

MA = pd.DataFrame([roll_boa,roll_citi,roll_gs,roll_jpm,roll_ms,roll_wfc],index=['BOA','Citi','GoldmanSachs','JPMorgan','MorganStanley','WellsFargo']).transpose()
MA
```

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2016**

```python
fig = plt.figure(figsize=(10,8))
MA['BOA'].plot(title='Rolling 30 day average VS Bank of America',legend = True)
BAC['close'].plot(legend = True)
```
